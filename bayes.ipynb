```python
def get_max_index(arr):
    max_i = 0
    max_v = arr[max_i]
    for i in range(1, len(arr)):
        if max_v < arr[i]:
            max_v = arr[i]
            max_i = i
    return max_i
```

```python
def get_disease_probs(disease):
    disease_probs = []
    all_p = disease.values[-1]
    for i in disease.values:
        if(i == all_p): continue
        disease_probs.append(i / all_p)
    return disease_probs
```

```python
def get_probs(disease, symptom, current_symptoms, disease_probs):
    our_probs = [1] * (len(disease)-1)
    for i in range(len(disease)-1):
        our_probs[i] *= disease_probs[i]
        for j in range(len(symptom) - 1):
            if current_symptoms[j] == 1:
                our_probs[i] *= float(symptom.iloc[j][i+1].replace(',', '.'))
    return our_probs
```

```python
def bayes(disease, symptom, current_symptoms):
    disease_probs = get_disease_probs(disease['количество пациентов'])
    our_probs = get_probs(disease['Болезнь'], symptom, current_symptoms, disease_probs)
    max_i = get_max_index(our_probs)
    return disease['Болезнь'][max_i]
```
